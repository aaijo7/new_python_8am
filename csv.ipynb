```python
# concat
df1 = pd.DataFrame({'sn':[1,2,3],
                    'name':['Ram', 'Shyam', 'Hari'],
                    'add':['Kathmandu', 'Bhaktapur', 'Lalitpur']})
df2 = pd.DataFrame({'sn':[4,5,6],
                    'name':['Akash', 'Baral', 'Harry'],
                    'add':['Kathmandu', 'Bhaktapur', 'Lalitpur']})
df = pd.concat([df1,df2],ignore_index=True)
df
```
